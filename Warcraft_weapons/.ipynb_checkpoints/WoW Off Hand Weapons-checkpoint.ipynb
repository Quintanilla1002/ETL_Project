# WoW Off-Hand Weapons

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the Off-Hand Weapons csv
oh_file = "Resources/Off Hand.csv"
oh_df=pd.read_csv(oh_file)
oh_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_oh_df=oh_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps", "mlespeed"]]
clean_oh_df.head()
#Rename Columns
oh_renamed=clean_oh_df.rename(columns={"name_enus": "Name",
                                       "quality": "Quality",
                                       "dmgmin1": "Minumum Damage",
                                      "dmgmax1": "Maximum Damage",
                                      "dps": "Damage Per Second",
                                      "mlespeed": "Weapon Speed (in Seconds)"})
oh_renamed
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
oh_renamed=oh_renamed.dropna(how='any')
oh_renamed.head()
```
