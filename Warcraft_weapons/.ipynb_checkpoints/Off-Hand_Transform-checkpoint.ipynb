# WoW Held In Off-Hand Items

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the Held In Off-hand csv
heldoh_file = "Resources/Held In Off-hand.csv"
heldoh_df=pd.read_csv(heldoh_file)
heldoh_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_heldoh_df=heldoh_df[["name_enus", "quality"]]
clean_heldoh_df.head()
#Rename Columns
held_oh_renamed=clean_heldoh_df.rename(columns={"name_enus": "Name",
                                               "quality": "Quality"})
held_oh_renamed
#drop NaN columns
held_oh_renamed=held_oh_renamed.dropna(how='any')
held_oh_renamed.head()
```
