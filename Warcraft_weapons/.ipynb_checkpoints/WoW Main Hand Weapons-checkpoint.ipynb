# WoW Main Hand Items

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the Main Hand Weapons csv
mh_file = "Resources/Main Hand.csv"
mh_df=pd.read_csv(mh_file)
mh_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_mh_df=mh_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps","mlespeed"]]
clean_mh_df.head()
#Rename Columns
mh_renamed=clean_mh_df.rename(columns={"name_enus": "Name",
                                       "quality": "Quality",
                                       "dmgmin1": "Minumum Damage",
                                      "dmgmax1": "Maximum Damage",
                                      "dps": "Damage Per Second",
                                      "mlespeed": "Weapon Speed (in Seconds)"})
mh_renamed
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
mh_renamed=mh_renamed.dropna(how='any')
mh_renamed.head()
```
