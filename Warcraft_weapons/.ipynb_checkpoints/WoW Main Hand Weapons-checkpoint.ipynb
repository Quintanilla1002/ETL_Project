# WoW Main Hand Items

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the Main Hand Weapons csv
mh_file = "Resources/Main Hand.csv"
mh_df=pd.read_csv(mh_file)
mh_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_mh_df=mh_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps","mlespeed"]]
clean_mh_df.head()
#Rename Columns
mh_renamed=clean_mh_df.rename(columns={"name_enus": "name",
                                       "quality": "quality",
                                       "dmgmin1": "min_damage",
                                      "dmgmax1": "max_damage",
                                      "dps": "dps",
                                      "mlespeed": "weapon_speed_in_seconds"})
mh_renamed
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
mh_renamed=mh_renamed.dropna(how='any')
mh_renamed.head()
```

```python
#Create Connection to PgAdmin
rds_connection_string = "postgres:postgres@localhost:5432/Warcraft Weapons"
engine = create_engine(f'postgresql://{rds_connection_string}')

#Verify table names
engine.table_names()
```

```python
#Load Data Frames in SQL Database
mh_renamed.to_sql(name='mh_weapons', con=engine, if_exists='append', index=False)
```
