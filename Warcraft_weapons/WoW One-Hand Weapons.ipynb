# WoW One-Hand Weapons

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the One-Hand Weapons csv
onehand_file = "Resources/One-Hand.csv"
onehand_df=pd.read_csv(onehand_file)
onehand_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_onehand_df=onehand_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps", "mlespeed"]]
clean_onehand_df.head()
#Rename Columns
onehand_renamed=clean_onehand_df.rename(columns={"name_enus": "name",
                                       "quality": "quality",
                                       "dmgmin1": "min_damage",
                                      "dmgmax1": "max_damage",
                                      "dps": "dps",
                                      "mlespeed": "weapon_speed_in_seconds"})
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
onehand_renamed=onehand_renamed.dropna(how='any')
onehand_renamed.head()
```

```python
#Create Connection to PgAdmin
rds_connection_string = "postgres:postgres@localhost:5432/Warcraft Weapons"
engine = create_engine(f'postgresql://{rds_connection_string}')

#Verify table names
engine.table_names()
```

```python
#Load Data Frames in SQL Database
onehand_renamed.to_sql(name='onehand_weapons', con=engine, if_exists='append', index=False)
```
