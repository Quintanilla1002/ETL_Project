# WoW One-Hand Weapons

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the One-Hand Weapons csv
onehand_file = "Resources/One-Hand.csv"
onehand_df=pd.read_csv(onehand_file)
onehand_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_onehand_df=onehand_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps", "mlespeed"]]
clean_onehand_df.head()
#Rename Columns
onehand_renamed=clean_onehand_df.rename(columns={"name_enus": "Name",
                                       "quality": "Quality",
                                       "dmgmin1": "Minumum Damage",
                                      "dmgmax1": "Maximum Damage",
                                      "dps": "Damage Per Second",
                                      "mlespeed": "Weapon Speed (in Seconds)"})
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
onehand_renamed=onehand_renamed.dropna(how='any')
onehand_renamed.head()
```
