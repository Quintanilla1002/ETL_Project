# WoW Ranged Weapons

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read the Ranged Weapons csv
ranged_file = "Resources/ranged.csv"
ranged_df=pd.read_csv(ranged_file)
ranged_df.head()
```

```python
#Clean the data frame to have only relevant stats/columns
clean_ranged_df=ranged_df[["name_enus", "quality", "dmgmin1", "dmgmax1", "dps", "rgdspeed"]]
clean_ranged_df.head()
#Rename Columns
ranged_renamed=clean_ranged_df.rename(columns={"name_enus": "name",
                                       "quality": "quality",
                                       "dmgmin1": "min_damage",
                                      "dmgmax1": "max_damage",
                                      "dps": "dps",
                                      "rgdspeed": "weapon_speed_in_seconds"})
#drop NaN columns as these are not weapons and irrelevant to the dataset if they do not have a damage value
ranged_renamed=ranged_renamed.dropna(how='any')
ranged_renamed.head()
```

```python
#Create Connection to PgAdmin
rds_connection_string = "postgres:postgres@localhost:5432/Warcraft Weapons"
engine = create_engine(f'postgresql://{rds_connection_string}')

#Verify table names
engine.table_names()
```

```python
#Load Data Frames in SQL Database
ranged_renamed.to_sql(name='ranged_weapons', con=engine, if_exists='append', index=False)
```
